```python
# Import necessary libraries
import pandas as pd
```

```python
# Read the raw dataset and make a copy dataframe
raw_df = pd.read_excel("../1_datasets/raw_data/UN_Comtrade_imports_dataset_raw.xlsx")
df = raw_df.copy()
df.head()
```

```python
# Display the shape of the DataFrame (number of rows and columns)
df.shape
```

```python
# Display information about the DataFrame, including column data types and non-null values
df.info()
```

```python
# Display the column names of the DataFrame
df.columns
```

```python
# Define a list of columns to be dropped
columns_to_drop = [
    "typeCode",
    "freqCode",
    "refPeriodId",
    "refMonth",
    "period",
    "reporterCode",
    "flowCode",
    "flowDesc",
    "partnerCode",
    "partnerDesc",
    "partner2Desc",
    "partnerISO",
    "partner2Code",
    "partner2Code",
    "partner2ISO",
    "classificationCode",
    "classificationSearchCode",
    "isOriginalClassification",
    "aggrLevel",
    "isLeaf",
    "customsCode",
    "customsDesc",
    "mosCode",
    "motCode",
    "motDesc",
    "qtyUnitCode",
    "qtyUnitAbbr",
    "isQtyEstimated",
    "altQtyUnitCode",
    "altQtyUnitAbbr",
    "altQty",
    "isAltQtyEstimated",
    "isNetWgtEstimated",
    "grossWgt",
    "isGrossWgtEstimated",
    "cifvalue",
    "fobvalue",
    "qty",
    "legacyEstimationFlag",
    "isReported",
    "isAggregate",
]
# Drop the specified columns from the DataFrame and create a new DataFrame
new_df = df.drop(columns=columns_to_drop)
```

```python
# Display the column names of the new DataFrame
new_df.columns
```

```python
# Display the shape of the new DataFrame
new_df.shape
```

```python
# Rename the columns for clarity
new_df.rename(
    columns={
        "refYear": "Year",
        "\tISO_Code": "Country_Code",
        "reporterDesc": "Country",
        "cmdCode": "Product_Code",
        "cmdDesc": "Product_Description",
        "netWgt": "Net_Weight_kg",
        "primaryValue": "Value_USD",
    },
    inplace=True,
)

# Display the first 5 rows of the DataFrame with the new column names
new_df.head()
```

```python
# Display the column names of the DataFrame after renaming
new_df.columns
```

```python
# check missing values
new_df.isna().sum()
```

```python
# fill missing values
new_df.fillna({"Net_Weight_kg": new_df["Net_Weight_kg"].mean()}, inplace=True)
```

```python
new_df.isna().sum()
```

```python
# check duplicate rows
new_df.duplicated().sum()
```

```python
# Random samples from new dataframe
new_df.sample(10)
```

```python
# Save the cleaned DataFrame to an Excel file
new_df.to_excel(
    "../1_datasets/cleaned_data/UN_comtrade_clean_dataset.xlsx",
    index=False,
)
```
